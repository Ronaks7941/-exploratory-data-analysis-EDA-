```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('/content/Real estate.csv')

# Display basic information about the dataset
print(data.info())

# Display summary statistics
print(data.describe())

# Check for missing values
print(data.isnull().sum())

# Visualize the distribution of the target variable
plt.figure(figsize=(10, 6))
sns.histplot(data['X4 number of convenience stores'], bins=20, kde=True)
plt.title('Distribution of Number of Convenience Stores')
plt.xlabel('Number of Convenience Stores')
plt.ylabel('Frequency')
plt.show()

# Visualize the relationships between features and the target variable
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Y house price of unit area', y='X4 number of convenience stores', data=data)
plt.title('House Price vs. Number of Convenience Stores')
plt.xlabel('House Price (Y house price of unit area)')
plt.ylabel('Number of Convenience Stores')
plt.show()

# Correlation matrix
correlation_matrix = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
